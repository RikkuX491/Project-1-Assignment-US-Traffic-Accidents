### Dataset Structure

* **Accident Severity**

* **Temporal Features**

* **Location and Geographic Information**

  * `Timezone`: Time zone associated with the location.

* **Weather and Environmental Conditions**

  * `Temperature(F)`, `Humidity(%)`, `Pressure(in)`, `Visibility(mi)`, `Wind_Speed(mph)`, `Weather_Condition`
  * These variables provide context about environmental factors that may influence accident occurrence and severity.

* **Roadway and Infrastructure Features**

  * Indicators such as `Amenity`, `Bump`, `Crossing`, `Give_Way`, `Junction`, `Traffic_Signal`, `Stop`, and `Roundabout`.
  * These binary variables describe nearby road features that may affect driving conditions and accident risk.

* **Distance and Impact Indicators**

  * `Distance(mi)`: Length of the road segment affected by the accident.
  * Additional contextual variables describing traffic disruptions and road conditions.

```python
# Identify Potential Outliers Using IQR

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]

    return len(outliers)

outlier_summary = {}

for col in numerical_cols:
    outlier_summary[col] = detect_outliers_iqr(df, col)

pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier Count']).sort_values(by='Outlier Count', ascending=False).head(10)
```

## Data Cleaning and Preprocessing

```python
# ==============================
# Create Copy of Dataset
# ==============================

df_clean = df.copy()

print("Initial Shape:", df_clean.shape)
```

## Convert Datetime Columns

```python
# ==============================
# Convert Time Columns to Datetime
# ==============================

df_clean['Start_Time'] = pd.to_datetime(df_clean['Start_Time'], errors='coerce')
df_clean['End_Time'] = pd.to_datetime(df_clean['End_Time'], errors='coerce')
```

## Feature Engineering — Time-Based Features

Temporal patterns are important for understanding accident trends.

```python
# ==============================
# Time-Based Feature Engineering
# ==============================

df_clean['Hour'] = df_clean['Start_Time'].dt.hour
df_clean['DayOfWeek'] = df_clean['Start_Time'].dt.day_name()
df_clean['Month'] = df_clean['Start_Time'].dt.month
df_clean['Year'] = df_clean['Start_Time'].dt.year

# Weekend indicator
df_clean['Is_Weekend'] = df_clean['DayOfWeek'].isin(['Saturday', 'Sunday']).astype(int)

# Accident duration in minutes
df_clean['Duration_Minutes'] = (
    (df_clean['End_Time'] - df_clean['Start_Time']).dt.total_seconds() / 60
)
```

## Handling Missing Values

Missing values are handled differently depending on the variable type and importance.

Strategy:

* Remove rows missing critical fields (location, severity, time).
* Impute numerical weather variables using median values.
* Fill categorical variables with “Unknown” where appropriate.

```python
# ==============================
# Remove Rows Missing Critical Fields
# ==============================

critical_cols = ['Severity', 'Start_Time', 'Start_Lat', 'Start_Lng']

df_clean = df_clean.dropna(subset=critical_cols)

print("Shape after dropping critical missing values:", df_clean.shape)
```

```python
# ==============================
# Impute Numerical Weather Columns
# ==============================

num_cols = [
    'Temperature(F)', 'Humidity(%)', 'Pressure(in)',
    'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)'
]

for col in num_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())
```

```python
# ==============================
# Fill Categorical Missing Values
# ==============================

categorical_cols = df_clean.select_dtypes(include='object').columns

for col in categorical_cols:
    df_clean[col] = df_clean[col].fillna('Unknown')
```

## Standardize Weather Condition Categories

Weather descriptions often contain many variations.

```python
# ==============================
# Simplify Weather Conditions
# ==============================

def simplify_weather(condition):
    condition = str(condition).lower()

    if 'rain' in condition:
        return 'Rain'
    elif 'snow' in condition:
        return 'Snow'
    elif 'fog' in condition or 'mist' in condition:
        return 'Fog/Mist'
    elif 'cloud' in condition:
        return 'Cloudy'
    elif 'clear' in condition or 'fair' in condition:
        return 'Clear'
    else:
        return 'Other'

df_clean['Weather_Category'] = df_clean['Weather_Condition'].apply(simplify_weather)
```

## Outlier Treatment

Extreme values may distort analysis. We apply reasonable filtering for key variables.

```python
# ==============================
# Remove Extreme Outliers
# ==============================

# Duration should be positive and reasonable (< 24 hours)
df_clean = df_clean[(df_clean['Duration_Minutes'] >= 0) &
                    (df_clean['Duration_Minutes'] <= 1440)]

# Visibility should be realistic
if 'Visibility(mi)' in df_clean.columns:
    df_clean = df_clean[df_clean['Visibility(mi)'] <= 100]

# Wind speed realistic range
if 'Wind_Speed(mph)' in df_clean.columns:
    df_clean = df_clean[df_clean['Wind_Speed(mph)'] <= 200]
```

## Boolean Feature Conversion

Many road features are stored as True/False.

```python
# ==============================
# Convert Boolean Columns to Integers
# ==============================

bool_cols = df_clean.select_dtypes(include='bool').columns

for col in bool_cols:
    df_clean[col] = df_clean[col].astype(int)
```

## Drop Irrelevant or High-Missing Columns (Optional)

Columns with excessive missing data or limited analytical value can be removed.

```python
# ==============================
# Drop Columns with High Missing Values
# ==============================

missing_threshold = 0.7  # 70%

missing_ratio = df_clean.isnull().mean()

cols_to_drop = missing_ratio[missing_ratio > missing_threshold].index

df_clean = df_clean.drop(columns=cols_to_drop)

print("Dropped columns:", list(cols_to_drop))
print("Final Shape:", df_clean.shape)
```
